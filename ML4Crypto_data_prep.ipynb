```python
from imageio import imread
import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('TrainingData.csv')
df.describe()
```

# creating an image from each ciphertext - an example

```python
a = df.iloc[0]['Text']
```

```python
len(a)
```

<b>Goal</b>: convert ciphertext into a square PGMA image. Convert this PGMA image to .png image

we collect the list of characters from ciphertext and obtain ascii values corresponding to their characters

Additionally we are appending 25 values each with a value 255 to the above list.

This list helps us in generating a square image of size 45 x 45 

```python
s = []
for i in a:
    s.append(str(ord(i)))
for i in range(25):
    s.append('255')
s = " ".join(s)
s = s+" "#this additional space is added to facilitate conversion from .pnm to .png which is later required by pnmtopng 
```

```python
#creating header for PGMA file. For details check: https://people.sc.fsu.edu/~jburkardt/data/pgma/pgma.html
f = open("new.pnm","w")
f.write("P2 45 45 255\n")
f.write(s)
f.close()
```

```python
#using pnmtopng utility to convert .pnm file to png file
# note that pgma is one of the sub classes of pnm file types
os.system("pnmtopng -compression 0  new.pnm > new.png")
```

```python
#viewing the so generated image
img = imread("new.png")
plt.figure()
plt.imshow(img,cmap='gray')
```

# Converting each ciphertext into images as described above. Labelling is done as per their class label

```python
ldf = len(df)
```

```python
path = !pwd
path = path[0]
path
```

```python
class_count = np.zeros(4)# counters for each class label, so that we use these counters to label images belonging to different classes
for i in range(ldf):
    txt = df.iloc[i]['Text']
    clabel = df.iloc[i]['class']
    s = []
    for i in txt:
        s.append(str(ord(i)))
    for i in range(25):
        s.append('255')
    s = " ".join(s)
    s = s+" "
    class_count[int(clabel)]+=1
    fname = str(clabel)+"_"+str(int(class_count[int(clabel)]))+".pgm"
    f = open(path+'/pgmdata/'+fname,"w")
    f.write("P2 45 45 255\n")
    f.write(s)
    f.close()
    os.system("pnmtopng -compression 0 "+path+'/pgmdata/'+fname+" > "+path+'/pngdata/'+fname[:-3]+"png")
```
